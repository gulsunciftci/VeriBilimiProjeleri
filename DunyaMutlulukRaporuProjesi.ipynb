```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("2015.csv")
df2 = pd.read_csv("2016.csv")
df3 = pd.read_csv("2017.csv")
df4 = pd.read_csv("2018.csv")
df5 = pd.read_csv("2019.csv")
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.head(20)
```

```python
df.columns
```

```python
scorehappiness_turkey=df.groupby("Country")["Happiness Score"].get_group("Turkey").reset_index()
```

```python
scorehappiness_turkey
```

```python
scorehappiness_countries=df.groupby("Country")["Happiness Score"].mean().reset_index()
```

```python
scorehappiness_countries
```

```python
scorehappiness_countries.sort_values("Happiness Score",ascending=False).head(20).plot(x="Country",y="Happiness Score",kind="bar")
```
